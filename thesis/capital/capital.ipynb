# 劳动价值论和边际价值理论

## 分析的角度看  

《资本论》即使处处透露着马克思的哲学思想，但是不管怎么说，抛开了她的经济学内涵不谈是一种逃避，毕竟她不是写给哲学家这类高端玩家。关于价值论，如今教育就像躲避瘟疫一样避而不谈，只接触它的最终形式：消费者行为理论。

商品价值是什么？奥地利学派信奉极端的个人主义，然而在这个问题上它的回答更容易让人接受。按照主观价值理论，乃至后来的边际价值理论的说法，物品为何有价值？因为它对人有用，带来人处境的改善，商品也是如此。人是不断做选择的动物，交换的达成只能是因为双方都愿意，双方都得到了处境的改善，不然就不会有交易。因而交易不是零和博弈，不必须是“等价交换”。相对的，马克思的劳动价值论认为，商品是用于交换的劳动产品，其价值由凝结在该商品中的无差别的人类劳动量决定。这两者差别在哪里？  

商品价值的来源是人类劳动，这是正确的。因为没有人的劳动痕迹的物品是纯自然的物品，不是商品。但是《资本论》通过商品的相对价值形式、等价形式等一列变换，最终推导出劳动价值论的做法值得商榷。因为这样一来，一般人类劳动虽然为衡量商品价值提供了** 量化方法** ，但是同时脱离了人自身。相反，主观效用虽然是不可度量的，但是人们追求满足自己需求的东西，即追求效用的满足不能算作是一种“异化”，即没有脱离人本身，没有和人相对立。  

换一种说法，我认为两种价值论的差别，开始只是源于看问题的角度不同，但是最终却引起了截然不同的结论。说严重些，就是屁股决定脑袋。主观价值论的角度是，将商品理解做满足人需求的一种存在，强调商品价值的** “去向”** ，这种“去向”直接回归人，没有拉开商品与人之间的距离，这是明显的个人主义观点；劳动价值论的角度是，将商品理解做劳动的产物这样一种存在，强调商品价值的** “来源”** ，而劳动其实是一种物质性活动即实践，于是这是一种唯物主义的观点，下面仔细说明这一点。  

劳动价值论和马克思的辩证唯物主义根本上是相通的。马克思认为物质决定意识，是说人们在追求效用满足时必定受制于物质条件，或者说规律。从实践角度来说就是，实践的目的性必须要服从实践的物质性。由于商品是用于交换的物质性活动的产品，而物质性支配着目的性，** 因此商品价值首要地体现在它的* “来源”* 性，即物质性人类劳动当中，之后才可能会体现在它的* “去向”* ，即人的效用满足上。**  

这样说来，在商品价值的决定过程中，是商品价值的** “来源”** 性占据了支配地位，而商品价值的** “去向”** 性只能占据从属地位了吗？其实我认为不是，因为我们忽略了讨论的前提：当我们讨论商品的价值怎么决定时，隐含的假定是商品已经生产出来了。这个生产出来的商品独立于实践的物质性、目的性等等属于生产过程的性质，也就是说当我们面前摆着一件商品时，它的价值只能由它的** “去向”** 性决定，即它能满足多少人的主观效用。因此它的价值因人而异，这不仅牢牢扣住回归人这个主题，而且符合常识。  

那么，商品价值的** “来源”** 性既然不能在决定商品价值这个地方发挥作用，那到底能在什么地方发挥作用？一句话，它在商品生产之前，决定了人与人之间的依赖关系。资本家凭借着手中的资本，事实上拥有了生产的条件，这是一种物质性力量，或者说自然的力量。生产这种实践活动必须依赖资本，而且是首要地依赖资本，其次才会产生效用的满足。  

## 经济学的角度看

我的观点，商品价值的** “来源”** 性决定了人与人之间的依赖关系，是有依据的。为何工人不能选择离开工厂，不受资本家剥削呢？正是由于这种依赖性，工人本身除了肉体之外一无所有，这是资本原始积累的后果。《资本论》中关于工人处境以及国家法令的描述很详尽，工人处境悲惨，而强制劳动的法令甚至对乞丐都有约束力，说明政府机构在原始积累的过程中扮演重要角色，而最终把工人关在工厂里的，根本上不是法令，而是生产之前就决定的人与人间的依赖关系。  

## 成本分析的角度  

这里是纯经济分析，如果说上面的分析是最近才产生的想法，这里的分析则老早就产生了。  
在《资本论》那里，原料成本作为物化的劳动，直接构成商品价值的一部分。而我认为成本根本不可能是价值的构成部分，相反，成本扮演的是“机会成本”的角色。** 不是成本决定了价格，而是价格限制了成本，**（完）
