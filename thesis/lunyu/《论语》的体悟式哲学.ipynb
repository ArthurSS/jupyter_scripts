# 摘要

《论语》作为一部记录夫子言行的作品，是儒家思想的集中反映，两千多年来深刻影响了中国人思想的方方面面，并在很大程度上奠定了中华文化的基调。经者，常也。历史上对《论语》的阐发一直没有停息，各种文献资料浩如烟海。然而毫无疑问的是，《论语》依旧具有鲜活的生命力，并将在未来继续塑造我们的民族特质。  
思想与知识是根本不同的两个范畴。学富五车的知识并不一定产生智慧，而一个文盲，知识仅限于生活经验的人，也可能有很高的智慧。在当下这个日新月异的世界，知识的更新速度快得令人难以想象，各种学科层出不穷，然而思想不能积累而只能阐发，忙碌的当代人在思想上已经不可能超越体天察地的古人。因此，《论语》伟大之处不是给予我们多少知识，而是把夫子的思想用一种现实而生动的形式表达了出来，就像一轮明月，千百年来面貌未改，却引发人们无数遐想，又如灿烂的星空，我们抬头仰望，顿生向往之情。  
《论语》透露出的是一种“体悟”式的哲学思想，将人的精神寄托内向地交付，而不是交给神明，这和西方“外向超越”的文化有本质的不同。虽然夫子有时会提到“天道”，然而道不远人（中庸），我欲仁，斯仁至矣（论语·述而）因此，探究这种体悟性而非逻辑性的思维，有助于我们把握中华文化的体质，从而更好地传承她。本文将从“体悟”的角度，结合中西方文化的比较，论证《论语》各个方面的思想在当代的现实意义，以及我们应该如何传承这种思想。  
关键词： 论语 哲学 内向超越 国学 传统文化
