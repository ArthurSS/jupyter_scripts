```python
# encoding:utf-8

import datetime
import pymongo
import pprint
from bson.objectid import ObjectId

import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt

myfont = mpl.font_manager.FontProperties(fname='C:/Windows/Fonts/msyh.ttf')  
mpl.rcParams['axes.unicode_minus'] = False

#创建服务器
default_client = pymongo.MongoClient('localhost', 27017)
#选择数据库
space_db = default_client['VnTrader_Space_Parity_Db']
#选择数据表
m1705_c_2450_coll = space_db['m1705-C-2450']
m1705_c_2500_coll = space_db['m1705-C-2500']
m1705_c_2550_coll = space_db['m1705-C-2550']
m1705_c_2800_coll = space_db['m1705-C-2800']
m1705_c_2850_coll = space_db['m1705-C-2850']
m1705_c_2900_coll = space_db['m1705-C-2900']
m1705_c_2950_coll = space_db['m1705-C-2950']
m1705_c_3000_coll = space_db['m1705-C-3000']
m1705_c_3050_coll = space_db['m1705-C-3050']
m1705_c_3100_coll = space_db['m1705-C-3100']
#选择文档
datetime1 = datetime.datetime(2016, 12, 28, 9, 10, 0)
datetime2 = datetime.datetime(2016, 12, 28, 15, 0, 0)
columns_ = ['close_ca', 'close_re', 'open_ca', 'open_re', 'target_ca', 'target_re', 'fore_ca',  'fore_re', 'time']
docs = m1705_c_2500_coll.find({'time':{'$gt':datetime1, '$lt':datetime2}}, columns_).sort([('datetime', pymongo.ASCENDING)])

result = pd.DataFrame(list(docs), columns=columns_)
result = result[(abs(result.fore_ca)<100)&(abs(result.fore_re)<100)].dropna(how='any')
result.index = result.loc[:, 'time']
print(result)
#for doc in docs:
#    series_ = pd.Series(doc, index=[doc['time']])
#    save_.append(series_)
```
