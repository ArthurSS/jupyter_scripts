```python
#encoding:utf-8
#10 Minutes to Pandas
#这里包含了10min2pandas的主要示例脚本
import math
import numpy as np
import pandas as pd

#1. Object Creation
dates = pd.date_range('20170501', periods=6)
df = pd.DataFrame(data = np.random.randn(6,4), index = dates, columns=list('ABCD'))
print('init: \n %r' % df)

#2. Sorting
df = df.sort_index(axis=1, ascending=False)
print('sorted: \n %r' % df)

#3. Setting
df.loc[dates[0],:] = [1,2,3,4]
#Add new column 'F'
dates2 = pd.date_range('20170503', periods=6)
s2 = pd.Series(5, index = dates2)
df['F'] = s2
print('new column added: \n %r' % df)

#4. Boolean Selection
print("the tpye of df.F is:%r" % type(df.F))
print('select via filter, df.loc[pd.isnull(df.F),:] \n %r' % df.loc[pd.isnull(df.F),:])
print('select via filter, df.loc[:,pd.isnull(df.loc[dates[0]])] \n %r' % df.loc[:,pd.isnull(df.loc[dates[0]])])

print('select via filter, df[df>0] \n %r' % df[df>0])
print('select via filter, df[pd.notnull(df)] \n %r' % df[pd.notnull(df)])

#5. apply, applymap, map
print("apply, applymap, map")
def call(num):
    def round_(x):
        return round(x, num)
    return round_

#DataFrame对象有apply, applymap, 没有map
#Series对象有apply, map, 没有applymap
#apply()将一个函数作用于DataFrame中的每个行或者列
#而applymap()是将函数做用于DataFrame中的所有元素(elements)
print(df.iloc[2].apply(call(2)))
print(df.iloc[2].map(call(2)))
print(df.applymap(call(2)))
print(df)
```

```python
# encoding:utf-8
# 这里包含了自己探索的一些语法细节
import numpy as np
import pandas as pd
import time
#position数据框
pos_col = ['longPos', 'shortPos', 'netPos', 'longAP', 'shortAP']
pos_index = ['m1703', 'm1703-C-3000', 'm1703-P-3000']
pos_df = pd.DataFrame(columns=pos_col, index=pos_index, data=np.array([[12]*len(pos_col)]*len(pos_index)))
print(pos_df)
print("----------------------------------------------------------")

#通过loc[]得到的选取结果都是dataframe类型的
print(u"通过loc[]得到的选取结果都是dataframe类型的")
print("type(pos_df.iloc[:,[0]]): %r" % type(pos_df.iloc[:,[0]]))
print("----------------------------------------------------------")

#通过interrows()得到的row是series类型的
#series选取特定index值的语法为 s1[['', '']]
print(u"通过interrows()得到的row是series类型的")
print(u"series选取特定index值的语法为 s1[['', '']]")
for index, row in pos_df.iterrows():
    print(row.to_dict())
    print(row[['longPos', 'shortPos']])
print("----------------------------------------------------------")

#两个series是可以比较的，两个dataframe也是可以比较的
print(u"两个series是可以比较的，两个dataframe也是可以比较的")
print("pos_df.iloc[:, 0]==pos_df.longPos: ")
print(pos_df.iloc[:, 0]==pos_df.longPos)
print("type(pd.isnull(pos_df.longPos)): %r" % type(pd.isnull(pos_df.longPos)))
print("----------------------------------------------------------")

#loc[]可以接受两个bool型的series作为参数
#loc[]当然也可以接受两个index作为参数
print(u"loc[]可以接受两个bool型的series作为参数")
print(u"loc[]当然也可以接受两个index作为参数")
index_bool = pd.Series([True, True, False], index=pos_index)
col_bool = pd.Series([True, False, True, False, True], index=pos_col)
print("index_bool = pd.Series([True, True, False], index=pos_index)")
print("col_bool = pd.Series([True, False, True, False, True], index=pos_col)")
print("pos_df.loc[index_bool, col_bool]: ")
print(pos_df.loc[index_bool, col_bool])

print(pos_df.loc[pos_df.index, pos_df.columns])
print("----------------------------------------------------------")

#df.index.values是np.array类型，然而series.values是list类型
print(u"df.index.values是np.array类型，然而series.values是list类型")
print(list(pos_df.index.values), type(pos_df.columns.values))
print(pos_df.longPos.values)
```
