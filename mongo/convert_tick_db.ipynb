```python
#encoding:utf-8
import datetime
import pymongo

#创建所有的表名
all_symbols = ["m1708-C-2600", "m1708-C-2650", "m1708-C-2700", "m1708-C-2750", "m1708-C-2800", "m1708-C-2850", "m1708-C-2900", "m1707-C-2500", "m1707-P-2500", "m1708-C-2500", "m1708-P-2500", "m1712-C-2550", "m1712-P-2550", "m1707-C-3150", "m1707-P-3150", "m1707-C-3100", "m1707-P-3100", "m1805", "m1805-P-2600", "m1805-P-2650", "m1805-C-2500", "m1805-C-2550", "m1805-C-2600", "m1805-C-2650", "m1805-C-2700", "m1805-C-2750", "m1805-C-2800", "m1805-C-2850", "m1805-C-2900", "m1805-C-2950", "m1805-P-2500", "m1805-P-2550", "m1711-C-3100", "m1711-P-3100", "m1805-P-2700", "m1805-P-2750", "m1805-P-2800", "m1805-P-2850", "m1805-P-2900", "m1805-P-2950", "m1709-C-3100", "m1709-P-3100", "m1801", "m1709-P-2600", "m1708-C-2950", "m1708-C-3000", "m1708-C-3050", "m1708-P-2600", "m1708-P-2650", "m1708-P-2700", "m1708-P-2750", "m1708-P-2800", "m1708-P-2850", "m1708-P-2900", "m1708-P-2950", "m1708-P-3000", "m1708-P-3050", "m1709-C-2550", "m1709-C-2600", "m1709-C-2650", "m1709-C-2700", "m1709-C-2750", "m1709-C-2800", "m1709-C-2850", "m1709-C-2900", "m1709-C-2950", "m1709-C-3000", "m1709-P-2550", "m1709-P-2650", "m1711-P-2900", "m1709-P-2700", "m1709-P-2750", "m1709-P-2800", "m1709-P-2850", "m1709-P-2900", "m1709-P-2950", "m1709-P-3000", "m1711-C-2600", "m1711-C-2650", "m1711-C-2700", "m1711-C-2750", "m1711-C-2800", "m1711-C-2850", "m1711-C-2900", "m1711-C-2950", "m1711-C-3000", "m1711-C-3050", "m1711-P-2600", "m1711-P-2650", "m1711-P-2700", "m1711-P-2750", "m1711-P-2800", "m1711-P-2850", "m1712-C-2500", "m1712-P-2500", "m1711-P-2950", "m1711-P-3000", "m1711-P-3050", "m1712-C-2650", "m1803-P-2800", "m1803-P-2850", "m1803-P-2900", "m1803-P-2950", "m1803-P-3000", "m1803-P-3050", "m1803-P-3100", "m1712-C-2700", "m1712-C-2750", "m1712-C-2800", "m1712-C-2850", "m1712-C-2900", "m1712-C-2950", "m1712-C-3000", "m1712-C-3050", "m1712-C-3100", "m1712-C-3150", "m1712-P-2650", "m1712-P-2700", "m1712-P-2750", "m1712-P-2800", "m1711", "m1712-P-2850", "m1712-P-2900", "m1712-P-2950", "m1712-P-3000", "m1712-P-3050", "m1712-P-3100", "m1712-P-3150", "m1801-C-2600", "m1801-C-2650", "m1801-C-2700", "m1801-C-2750", "m1801-C-2800", "m1801-C-2850", "m1801-C-2900", "m1801-C-2950", "m1801-C-3000", "m1801-C-3050", "m1801-P-2600", "m1801-P-2650", "m1801-P-2700", "m1801-P-2750", "m1801-P-2800", "m1801-P-2850", "m1801-P-2900", "m1801-P-2950", "m1712", "m1801-P-3000", "m1801-P-3050", "m1803-C-2650", "m1803-C-2700", "m1803-C-2750", "m1803-C-2800", "m1803-C-2850", "m1803-C-2900", "m1803-C-2950", "m1803-C-3000", "m1803-C-3050", "m1803-C-3100", "m1803-P-2650", "m1803-P-2700", "m1803-P-2750", "m1707-C-2600", "m1707-C-2650", "m1707-C-2700", "m1707-C-2750", "m1707-C-2800", "m1707-C-2850", "m1707-C-2900", "m1707-C-2950", "m1707-C-3000", "m1707-C-3050", "m1707-P-2600", "m1707-P-2650", "m1707-P-2700", "m1707-P-2750", "m1707-P-2800", "m1707-P-2850", "m1707-P-2900", "m1707-P-2950", "m1707-P-3000", "m1707-P-3050", "m1803", "m1708-C-3100", "m1708-P-3100", "m1709-C-3050", "m1709-P-3050", "m1801-C-3100", "m1801-P-3100", "m1707-C-2550", "m1707-P-2550", "m1712-C-2600", "m1712-P-2600", "m1803-C-2600", "m1803-P-2600", "m1709", "m1708-C-2550", "m1708-P-2550", "m1709-C-2500", "m1709-P-2500", "m1711-C-2550", "m1711-P-2550", "m1801-C-2550", "m1801-P-2550", "m1803-C-2550", "m1803-P-2550", "m1707", "m1708", "m1805-C-3000", "m1805-P-3000", "m1803-C-2500", "m1803-P-2500"]
all_symbols.sort()

#创建今天的三个时间段start_datetime和end_datetime
now = datetime.datetime.now()
start_datetime1 = datetime.datetime(now.year, now.month, now.day, 9, 10, 0, 0)
end_datetime1 = datetime.datetime(now.year, now.month, now.day, 10, 15, 0)
start_datetime2 = datetime.datetime(now.year, now.month, now.day, 10, 30, 0, 0)
end_datetime2 = datetime.datetime(now.year, now.month, now.day, 11, 30, 0)
start_datetime3 = datetime.datetime(now.year, now.month, now.day, 13, 30, 0, 0)
end_datetime3 = datetime.datetime(now.year, now.month, now.day, 15, 0, 0)

#获取时间datetime_list
time_interval = 500000 #时间间隔为0.5s
datetime_list1 = map(lambda x: start_datetime1+datetime.timedelta(microseconds=time_interval)*x, range(1000000*60*75/time_interval))
datetime_list2 = map(lambda x: start_datetime2+datetime.timedelta(microseconds=time_interval)*x, range(1000000*60*60/time_interval))
datetime_list3 = map(lambda x: start_datetime3+datetime.timedelta(microseconds=time_interval)*x, range(1000000*60*90/time_interval))


#构建查询条件
time_query = {
    '$or':[{'datetime':{'$gte':start_datetime1, '$lte':end_datetime1}},
    {'datetime':{'$gte':start_datetime2, '$lte':end_datetime2}},
    {'datetime':{'$gte':start_datetime3, '$lte':end_datetime3}}
    ]
}

#所有需要填充的字段
fill_key = [u'lastPrice', u'lowerLimit', u'openInterest', u'upperLimit', u'volume']

#获取数据库
default_client = pymongo.MongoClient('localhost', 27017)
tick_db = default_client['VnTrader_Tick_Db']
new_tick_db = default_client['tick_half_second']

def modify_record(record, modify_key='_id', id_info='', count_info=None):
    record[modify_key] = id_info + ' ' + str(count_info) + \
        datetime.datetime.strftime(datetime.datetime.now(), "%Y%m%d %H:%M:%S.%f")
    return record

for single_symbol in all_symbols:
    
    #获取数据表
    #给定合约名称，给定时间
    old_collection = tick_db[single_symbol]
    new_collection = new_tick_db[single_symbol]

    #获取指定时间段所有的纪录
    record_list = list(old_collection.find(time_query))
    #获取时间datetime_list
    datetime_list = datetime_list1 + datetime_list2 + datetime_list3
    
    if not record_list:
        print(u"执行价为%s的表没有查询到数据，跳过迁移" % single_symbol)
        continue
    #获取一个空白记录，作为填充使用
    blank_record = record_list[0]
    blank_record = {key: None for key in blank_record.keys() if key != '_id'}
    
    #cache_record作为上一次储存时的记录
    cache_record = record_list.pop(0)

    fill_count = 0
    time_point_count = 0
    condition1_count = 0
    condition2_count = 0
    condition3_count = 0
    inner_loop_count = 0
    inner_break = False
    #处理、储存部分开始
    #需要保证cache数据是完整tick，如果是做市商tick需要填充
    print(u"开始转化数据表: %s 长度为 %d" % (single_symbol, len(record_list)))
    while datetime_list:
        time_point_count += 1
        time_point = datetime_list.pop(0)
        #如果记录已经遍历完
        if not record_list:
            cache_record = modify_record(cache_record, id_info='cache record', count_info=time_point_count)
            new_collection.insert_one(cache_record)
            continue

        next_record = record_list[0]
        #如果上一条记录的时间没有超过时间点，而下一条记录的时间超过了储存时间点
        #则再次储存上一条记录，其他不变
        if cache_record['datetime'] <= time_point and next_record['datetime'] > time_point:
            condition1_count += 1
            cache_record = modify_record(cache_record, id_info='cache record', count_info=time_point_count)
            new_collection.insert_one(cache_record)
        #---------------------------------------------------------------------------------------
        #如果上一条和下一条记录的时间都超过时间点，该时间点之前还没有数据
        #故插入空白数据，其他不变
        elif cache_record['datetime'] > time_point and next_record['datetime'] > time_point:
            condition2_count += 1
            blank_record = modify_record(blank_record, id_info='blank record', count_info=time_point_count)
            new_collection.insert_one(blank_record)
        #---------------------------------------------------------------------------------------
        #如果上一条记录的时间没有超过时间点，而下一条记录的时间也没有超过时间点
        #则储存下一条数据，并且将下一条数据作为cache_record，并且在records中删除掉这条记录
        elif cache_record['datetime'] <= time_point and next_record['datetime'] <= time_point:
            condition3_count += 1
            #next_record值不变，但是删除了下一条记录
            next_record = record_list.pop(0)
            
            #正常情况下while语句不应该得到执行
            while record_list and record_list[0]['datetime'] <= time_point:
                inner_loop_count += 1
                next_record = record_list.pop(0)
            #如果即将储存的tick是做市商报价，而上一次储存的tick是完整tick，则填充即将储存的tick
            if next_record['volume'] == 0 and cache_record['volume'] != 0:
                for key in fill_key: next_record[key]=cache_record[key] 
                fill_count += 1 #填充次数加一
            next_record = modify_record(next_record, id_info='next record', count_info=time_point_count)
            new_collection.insert_one(next_record)
            cache_record = next_record
        else:
            raise Exception(u"条件判断出错")

    print(u"共储存时间点共%d个" % time_point_count)
    print(u"其中，三种条件分别执行次数为：%d, %d, %d" % (condition1_count, condition2_count, condition3_count))
    print(u"一共填充了%d次" % fill_count)
    print(u"内循环一共执行了%d次" % inner_loop_count)
    print(u"\n")
```
