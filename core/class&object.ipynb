```python
#新式类
class A(object):
    def __init__(self):
        pass
    def save(self):
        print "This is from A"
class B(A):
    def __init__(self):
        pass
class C(A):
    def __init__(self):
        pass
    def save(self):
        print  "This is from C"
class D(B,C):
    def __init__(self):
        pass
fun =  D()
fun.save()
#在新式类中，__class__属性返回自身类型，和type()返回的是同一个id()
#新式类的type是type对象
print("D.__class__ is type(D): %r, and type(D) is: %r" % ((D.__class__ is type(D)), type(D)))
#新式类实例的type是class

print("fun.__class__ is type(fun): %r, and type(fun) is: %r" % ((fun.__class__ is type(fun)), type(fun)))
print("the type of A, B, C, D is: %r, %r, %r, %r" % (type(A), type(B), type(C), type(D)))
print("#"*100)


#旧式类
#新式类
class A():
    def __init__(self):
        pass
    def save(self):
        print "This is from A"
class B(A):
    def __init__(self):
        pass
class C(A):
    def __init__(self):
        pass
    def save(self):
        print  "This is from C"
class D(B,C):
    def __init__(self):
        pass
fun =  D()
fun.save()
#旧式类的type是classobj, 而classobj的类型才是type
#旧式类的实例的type是instance, 而instance的类型也是type
print("type(D) is: %r" % type(D))
print("type(type(D)) is: %r" % type(type(D)))
print("type(fun) is: %r" % type(fun))
print("type(type(fun)) is: %r" % type(type(fun)))


print('a' > 'b')
print(type(object))
class Z():
    "this is doc"
    def __nonzero__(self):
        return False
    def __len__(self):
        return False
z = Z()
print Z.__dict__
print(bool(Z))
print(bool(z))
print(type(2**30+2**29+2**28))
a = 'aa' * 20
```
